## Business Dataset

```python
business = pd.read_json("yelp_dataset~\\yelp_academic_dataset_business.json", lines = True)
business.drop(axis = 1, columns = ['hours','address'], inplace = True)
business.dropna(axis = 0, how = 'any', inplace = True)
str_cols = ['name','attributes','city','categories']
business[str_cols] = business[str_cols].apply(func = lambda x: x.str.strip().str.lower(), axis = 1)
restaurant = business.loc[business['categories'].str.contains('restaurant'),:].reset_index(drop = True)
restaurant.rename(columns = {'state':'code'}, inplace = True)
state_abb = pd.read_csv('State Abbreviations csvData.csv', usecols = ['State','Code'])
state_abb.rename(columns = {'State':'state', 'Code':'code'}, inplace = True)
restaurant = pd.merge(left = restaurant, right = state_abb, how = 'left', on = 'code')
restaurant.head(1)
```

```python
data = pd.merge(left = restaurant, right = reviews, how = 'inner', on = 'business_id')
data.head()
```

```python
category = dict()
for c in data['categories']:
    for x in c.split(','):
        if x.strip() in category.keys():
            category[x.strip()] = category[x.strip()] + 1
        else:
            category[x.strip()] = 1
category
```

```python
state = dict()
for s in data['state']:
    if s in state.keys():
        state[s] = state[s] + 1
    else:
        state[s] = 1
state
```

```python
city = dict()
for c in data['city']:
    if c in city.keys():
        city[c] = city[c] + 1
    else:
        city[c] = 1
city
```
